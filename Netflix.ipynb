```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset (corrected file path)
file_path = "C://Users//mlawal//Downloads//Dissertation Papers//Netflix//most-popular.csv"
data = pd.read_csv(file_path)
```

```python
# Setting the aesthetics for the plots
sns.set(style="whitegrid")

# 1. Category Analysis
plt.figure(figsize=(12, 6))
sns.countplot(data=data, y='category')
plt.title('Number of Shows/Movies in Each Category')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x='hours_viewed_first_91_days', bins=30, kde=True)
plt.title('Distribution of Hours Viewed in First 91 Days')
plt.xlabel('Hours Viewed')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='runtime')
plt.title('Distribution of Runtime Across Shows/Movies')
plt.xlabel('Runtime (Hours)')
plt.ylabel('Density')
plt.show()
```

```python
top_shows = data.sort_values(by='hours_viewed_first_91_days', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(data=top_shows, x='hours_viewed_first_91_days', y='show_title')
plt.title('Top 10 Shows/Movies Based on Hours Viewed in First 91 Days')
plt.xlabel('Hours Viewed')
plt.ylabel('Show Title')
plt.show()
```

```python
category_counts = data['category'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Different Categories')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='runtime', y='category')
plt.title('Distribution of Runtime Across Different Categories')
plt.xlabel('Runtime (Hours)')
plt.ylabel('Category')
plt.show()
```
